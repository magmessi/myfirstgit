```python
import requests
from bs4 import BeautifulSoup
import csv
from selenium import webdriver
import time
from itertools import zip_longest
import pandas as pd
```

```python
url = 'https://www.basketball-reference.com/leagues/NBA_2022_per_game.html'
```

```python
reponse = requests.get(url)

src = reponse.content

soup = BeautifulSoup(src, 'html.parser')

soup
```

```python
print(soup.prettify)
```

```python
tableau = soup.find_all(class_='full_table')
tableau[1]
```

```python
joueur =[]
for j in tableau[1].find_all('td'):
    print(j)
    joueur.append(j.text)
print(joueur)
```

```python
joueurs =[]

for i in range(len(tableau)):
    joueur = []
    for j in tableau[i].find_all('td'):
        joueur.append(j.text)
    joueurs.append(joueur)
print(joueurs)
```

```python
head= soup.find(class_='thead')
```

```python
column_name = [head.text for item in head][0]
column_name
                    
```

```python
column_name_clean = column_name.split("\n")[2:-1]
column_name_clean
```

```python
df = pd.DataFrame(joueurs,columns=column_name_clean)
df
```

```python
df = pd.DataFrame(joueurs,columns=column_name_clean).set_index('Player')
df
```
